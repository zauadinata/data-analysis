IMPORT PANDAS

```python
# Pandas is a library in Python used for data manipulation and analysis, primarily focused on tabular data.
import pandas as pd
```

READING FILE DATA

```python
# import file csv
df = pd.read_csv ('/content/dataset_simulasi - Sheet1 (1).csv')
```

```python
# displays file contents
print (df)
```

CHECKING DATA

```python
# displays the last 2 lines
print (df.tail (2))
```

```python
# displays basic statistics
print (df.describe ())
# can also use df.describe ()
```

```python
# displays summary information about the data frame
df.info()
```

```python
# displays the number of rows and columns in a data file
print (df.shape)
```

```python
# displays the column name
print (df.columns)
```

```python
# displays column data types
print (df.dtypes)
```

```python
# check empty data
df.isnull().sum()
```

```python
# check duplicate data
df.duplicated().sum()
```

SELECTION AND FILTER DATA

```python
# choose columns

# 1 column
df ['nama']

# multiple columns
df [['nama', 'gender', 'nilai']]
```

```python
# choose rows
df.iloc[0] # first row
df.iloc[0:5] # first 5 rows
df.iloc[-5:] # last 5 rows
```

```python
df.loc[0] # row with index 0
```

```python
# filter data
df [df['umur'] >= 24]
```

```python
# multiple filter data
df[(df['umur'] >= 24) & (df['nama'] == 'rangga')]
```

CHANGE DATA

```python
# add name column
df ['kelas'] = [2, 3,3,3,3,3,3,3,3,3]
print (df)
# if the previous row of data is 10,then the added data must be 10
```

```python
# delete name column
df = df.drop (columns = ['kelas'])
print (df.head (5))
```

```python
# change name column
df = df.rename (columns = {'nilai': 'ipk'})
print (df.head (5))
```

```python
# fill in blank data
df = df.fillna (0, inplace = True) # fill Nan with 0
df.dropna (inplace = True) # Remove rows containing NaN
```

SORTING AND GROUPING

```python
# sorting data

# sort data from largest to smallest
df.sort_values(by='umur', ascending= False)

# if sort data form smallest to largest change False to True
```

```python
# grouping data
df.groupby('nama')['umur'].mean()
```

PIVOT TABLE

```python
# summarize data with aggregation calculations such as average, sum, etc.
df.pivot_table(index='nama', values='umur', aggfunc='mean')
```

CROSSTAB

```python
# tabel frequency
pd.crosstab(df['nama'], df['umur'])
```

VISUALIZATION

```python
df['umur'].plot(kind='bar')
df.plot(kind='line', x='nama', y='umur')
df.hist(column='umur')
```

SAVE DATA

```python
df.to_csv('data.csv', index = False)
```
