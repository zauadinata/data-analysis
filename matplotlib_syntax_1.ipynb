```python
import matplotlib.pyplot as plt
```

MEMBUAT GRAFIK GARIS SEDERHANA

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Membuat plot
plt.plot(x, y)

# Menampilkan grafik
plt.show()
```

MENAMBAHKAN JUDUL DAN LABEL

```python
plt.plot(x, y)
plt.title("Grafik Garis")         # Menambahkan judul
plt.xlabel("Sumbu X")             # Label sumbu X
plt.ylabel("Sumbu Y")             # Label sumbu Y
plt.grid(True)                    # Menampilkan grid
plt.show()
```

MENGUBAH WARNA DAN GAYA GARIS

```python
plt.plot(x, y, color="red", linestyle="--", marker="o")  # Warna merah, garis putus-putus, dan marker bulat
plt.grid(True)
plt.show()
```

MEMBUAT SCATTER PLOT

menampilkan hubungan antar 2 variabel dalam bentuk titik di bidang koordinat.

```python
plt.scatter(x, y, color="blue")
plt.grid(True)
plt.show()
```

MEMBUAT BAR CHART

membandingkan nilai antar variabel yang berbeda

```python
plt.bar(x, y, color=['red', 'blue', 'orange', 'purple', 'cyan'])
plt.grid(axis='y')  # Menambahkan grid horizontal
plt.show()
```

MEMBUAT PIE CHART

menampilkan proporsi atau persentase dari total data

```python
# Data yang digunakan
labels = ['A', 'B', 'C', 'D', 'E']  # Label untuk setiap bagian
values = [2, 4, 6, 8, 10]  # Menggunakan nilai dari y

# Membuat pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%', colors=['red', 'blue', 'green', 'purple', 'orange'])

# Menampilkan grafik
plt.show()
```
