```python
# pandas i# Pandas is a library in Python used for data manipulation and analysis, primarily focused on tabular data.
import pandas as pd
```

```python
# creating data frame using dictionary
data = {
    'nama' : ['Andi', 'Budi', 'Caca', 'Sintia', 'Dafina'],
    'umur' : [22, 23, 30, 22, 22],
    'kota' : ['Jakarta', 'Palembang', 'Surabaya', 'Palembang', 'Jakarta']
}
```

```python
df1 = pd.DataFrame(data)
```

```python
print (df1)
```

MERGE ( )

```python
# merges dataframe
df2 = pd.DataFrame({
    'nama' : ['Andi', 'Budi', 'Caca', 'Sintia', 'Dafina'],
    'ID' : [1, 2, 3, 4, 5]

})

df_merged = pd.merge(df1, df2, on = 'nama', how = 'inner')
print (df_merged)
```

JOIN ( )

```python
# combining by index
df1 = pd.DataFrame({'nama' : ['Andi', 'Budi', 'Caca', 'Sintia', 'Dafina']}, index = [1, 2, 3, 4, 5])
df2 = pd.DataFrame({'ID' : [1, 2, 3, 4, 5]}, index = [2, 3, 4, 5, 6])

# using syntax join
df_joined = df1.join(df2, how="left")
print(df_joined)
```

CONCAT ( )

```python
# merges dataframe horizontally
df1 = pd.DataFrame({'nama' : ['Andi', 'Budi'],  'umur' : [22, 23]})
df2 = pd.DataFrame({'nama' : ['Caca', 'Sintia'],  'umur' : [30, 22]})

df_concat = pd.concat([df1, df2], axis = 1)
print (df_concat)
```

```python
# merges dataframe vertically
df1 = pd.DataFrame({'nama' : ['Andi', 'Budi'],  'umur' : [22, 23]})
df2 = pd.DataFrame({'nama' : ['Caca', 'Sintia'],  'umur' : [30, 22]})

df_concat = pd.concat([df1, df2], ignore_index=True)
print(df_concat)
```
